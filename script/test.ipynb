```python
from scipy.stats import binned_statistic as bin_mean
def binning(X,Y,bin = 10):
    mean,edges,_ = bin_mean(X,Y,bins=bin)
    Y_ret = mean
    X_ret = edges[1::]

    return X_ret, Y_ret
```
